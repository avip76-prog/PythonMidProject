```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. יצירת תיבת הבחירה (Dropdown)
dropdown = widgets.Dropdown(
    options=sorted(df['main_category'].unique().tolist()),
    description='בחר קטגוריה:',
    style={'description_width': 'initial'}
)

# 2. אזור התצוגה של התוצאות
output = widgets.Output()

# 3. הפונקציה שרצה בכל פעם שבוחרים קטגוריה
def show_category_stats(change):
    with output:
        clear_output() # מנקה את התוצאה הקודמת
        category = change['new'] # הקטגוריה שנבחרה
        
        # חישובים (על בסיס ה-df שלך)
        stats = df[df['main_category'] == category]
        avg_rating = stats['rating'].mean()
        count = len(stats)
        
        # הדפסה מעוצבת
        print(f"📊 סיכום עבור קטגוריית: {category}")
        print("-" * 30)
        print(f"⭐ דירוג ממוצע: {avg_rating:.2f}")
        print(f"📦 כמות מוצרים: {count}")
        
        # בונוס: מציג את 3 המוצרים הכי טובים בקטגוריה
        print("\n🏆 הטופ 3 בקטגוריה:")
        display(stats.nlargest(3, 'rating')[['product_name', 'rating']])

# חיבור התיבה לפונקציה
dropdown.observe(show_category_stats, names='value')

# הצגת הכל על המסך
display(dropdown)
display(output)
```
